```python
def KMeansClustering(X):
    wcss = []
    ks = []

    for i in range(1,11):
        kmeans = KMeans(n_clusters=i,init="k-means++",random_state=42,n_init=10)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)
        ks.append(i)

    # for i in range(len(ks)):
    #     print(ks[i])
    plt.plot(ks,wcss,'bx-')
    plt.title("KMeans using elbow method")
    plt.xlabel("K value")
    plt.ylabel("WCSS")
    
```

```python
KMeansClustering(X)
```
